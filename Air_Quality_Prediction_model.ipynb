**Handling the missing Values**

Convert all the -200 values to NaN

Replace all NaN values with mean of that specific column

```python
air_quality_data = air_quality_data.replace(to_replace=-200, value = np.NaN)
```

```python
air_quality_data.tail()
```

**The error you encountered is caused by the presence of time values in the DataFrame. The mean() function in pandas doesn't know how to handle time values when calculating the mean.
To resolve this issue, you need to separate the numeric columns from the non-numeric columns (like 'Date' and 'Time') and then calculate the mean for the numeric columns.**

**Identify the numeric columns in the DataFrame:**

```python
numeric_cols = air_quality_data.select_dtypes(include=['number']).columns
```

**Calculate the mean of the numeric columns:**

```python
numeric_mean = air_quality_data[numeric_cols].mean()
print(numeric_mean)
```

```python
#relacing the missing value with the mean value
air_quality_data = air_quality_data.fillna(numeric_mean.mean())
```

**Forecasting with FB Prophet Algorithm**

Developed by Facebook's Core Data Science team

Prophet is a type of forecasting model that is based on an **additive model** where non-linear trends are fit with yearly, weekly, and daily seasonality, plus holiday effects. It is best used for time series that have strong seasonal effects and several years of historical data.
We use this forecasting model because it is **easy to use, accurate, and fast**. It is also fully automatic, so you do not need to have any prior experience with forecasting to use it.

FB Prophet Documentation: https://facebook.github.io/prophet/docs/quick_start.html

```python
# converting the data from DD/MM/YYYY to YYYY-MM-DD
date_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(date_info)
```

```python
time_info = air_quality_data['Time']
print(time_info)
```

```python
time_info = time_info.apply(lambda x:x.replace('.',':'))
print(time_info)
```

```python
print(type(date_info))
print(type(time_info))
```

```python
# combining 2 series to a pandas dataframe
date_time = pd.concat([date_info, time_info], axis=1)
date_time.head()
```

```python
from prophet import Prophet
```

```python
#training the Prophet model
model = Prophet(yearly_seasonality=True)
model.fit(data)
```

```python
future = model.make_future_dataframe(periods=365, freq='H')
future.tail()
```

```python
forecast = model.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```

```python
fig1 = model.plot(forecast)
```

```python
fig2 = model.plot_components(forecast)
```
